```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\LENOVO\Downloads\Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
print(plt.style.available)
plt.style.use('grayscale')
```

```python
#Lineplot
```

```python
df.plot(kind = 'line', figsize = (10,5), title = 'Ice Cream Ratings', xlabel = 'Daily Ratings', ylabel = 'Scores')
plt.legend(loc = 'upper left', fontsize = 'small')
```

```python
#Barplot
```

```python
df.plot.barh(stacked = True, figsize = (20,4))
```

```python
df.plot.scatter(x = 'Flavor Rating', y = 'Overall Rating', s = 70, c = 'Red', figsize = (5,3), title = 'Relationship between Flavor and Overall Ratings')
```

```python
df.plot.hist(bins = 25, figsize = (5,5))
```

```python
df.boxplot()
```

```python
#Area Chart
```

```python
df.plot.area(figsize = (8,3))
plt.legend(loc = 'upper left', fontsize = 'small')
```

```python
df.plot.pie(y = 'Overall Rating', figsize = (8,8))
plt.legend(loc = 'upper right')
```
